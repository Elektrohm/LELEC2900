```python
# function to plot responses don't pay attention to it, it comes from make graph of 4th session
def plot_response(fs, w, h, zoom=False):
    fig, ax1 = plt.subplots()
    ax1.set_title('Digital filter frequency response')
    ax1.plot(0.5*fs*w/np.pi, 20*np.log10(np.abs(h)))
    ax1.set_ylabel('Amplitude [dB]', color='steelblue')
    ax1.set_xlabel('Frequency [Hz]', color='steelblue')
    ax2 = ax1.twinx()
    angles = np.unwrap(np.angle(h))
    ax2.plot(0.5*fs*w/np.pi, angles, 'crimson')
    ax2.set_ylabel('Angle (radians)', color='crimson')
    ax2.grid()
    ax2.axis('tight') 
    if zoom :
        ax3    = plt.axes([1.2, 0.6, .45, .25])
        x      = 0.5*fs*w/np.pi
        signal = 20*np.log10(np.abs(h))
        ax3.plot(x, signal)
        ax3lim = [1500,2000]
        ax3.set_xlim(ax3lim[0], ax3lim[1])
        ax3.set_ylim(np.min(signal[(x>ax3lim[0]) & (x<ax3lim[1])]), 
                     np.max(signal[(x>ax3lim[0]) & (x<ax3lim[1])]))
        ax3.grid(True)

        ax4    = plt.axes([1.2, 0.25, .45, .25])
        x      = 0.5*fs*w/np.pi
        signal = 20*np.log10(np.abs(h))
        ax4.plot(x, signal)
        ax4lim = [2500,3000]
        ax4.set_xlim(ax4lim[0], ax4lim[1])
        ax4.set_ylim(np.min(signal[(x>ax4lim[0]) & (x<ax4lim[1])]), 
                     np.max(signal[(x>ax4lim[0]) & (x<ax4lim[1])]))
        ax4.grid(True)
    
    plt.show()
```

```python
##########################################################################
# TO COMPLETE (Question 2)
##########################################################################

#insert your code here. 
import scipy.signal as signal 

# fp : start of pass band, fstop : start of stop band
fp = 500
fstop = 1600

# width of the pass band normalize to nyquist freq, allow 0.005 ripple
width = (fstop-fp)/(0.5*Fs)
ripple = -20*np.log10(0.005)

# compute order for kaiser window 
order, beta = signal.kaiserord(ripple, width)
# ensure order is odd value
order |= 1
print(f"\nNeed order {order}")

# compute band pass FIR filter coefficient and impulse response
coef = signal.firwin(order, [fp, fstop], window=("kaiser",beta), fs=Fs, scale=False, pass_zero=False)
w2, h2 = signal.freqz(coef, worN=2000)
plot_response(Fs, w2, h2, 'lowpass')

# compute the filtered signal 
audioSeqFiltered = np.convolve(audio, coef, mode="same")

# compute fft of filtered signal 
fft_filtered = np.fft.fft(audioSeqFiltered)
fftfreq_filtered = np.fft.fftfreq(len(audioSeqFiltered), d=1/Fs)

# plot time and frequency signals
plotSignalAndDft(time, audioSeqFiltered, fftfreq_filtered, np.abs(fft_filtered), x_label=['time [s]','frequency [Hz]'], y_label=['',''], discrete=[True, True])
```

<font size=6 color=#009999> 3. Perform DFT of each segment to identify tones  </font> <br>
<br>

It is now the time to identify the digit that has been dialed in a segment based on the analysis of its frequency content. Most of the code has been written for you in the following cell (understanding functions inputs and outputs is sufficient to answer to the question). 

```python
##########################################################################
# TO RUN (some helper functions)
##########################################################################

from scipy.signal import find_peaks

"""--------------------------------------------------------------------------------------------------
Find the location of the 2 maxima of the "DftAmplitude" and return the corresponding frequencies in 
ascending order. 
INPUTS: 
    - DftAmplitude : 1-D signal with the aplitude of the DFT of a segment. 
    /!\ ONLY POSITIVES FREQUENCIES - Do not pass the entire spectrum but only the positive part. 
    - FreqAxis : 1-D signal (same shape than DftAmplitude) with the frequency axis of the DFT.
OUTPUT: 
    - Array with the frequencies showing the 2 highest amplitudes.
--------------------------------------------------------------------------------------------------"""
def findTones(DftAmplitude, FreqAxis): 
    peaks, _    = find_peaks(DftAmplitude)
    Z           = [x for _,x in sorted(zip(DftAmplitude[peaks],peaks),reverse=True)]
    lowF, highF = min(FreqAxis[Z[0]], FreqAxis[Z[1]]), max(FreqAxis[Z[0]], FreqAxis[Z[1]])
    return [lowF, highF]

"""--------------------------------------------------------------------------------------------------
Given the two frequency component of a segment, identifyDigits find the most likely digit that have 
been dialed. 
INPUTS: 
    - foundPeaks : output of the findTones function. 
    - tones : the "tones" dictionnay defined in the first cell (do not modify it).
OUTPUT: 
    - the most likely digit that has been dialed given the two extracted frequencies.
--------------------------------------------------------------------------------------------------"""
def identifyDigits(foundPeaks, tones): 
    assert len(foundPeaks) == 2, "each foundPeaks array sould contain two elements"
    lowFreq, highFreq         = foundPeaks[0], foundPeaks[1]
    identifiedDigit, errorMin = '?', 10000000
    for number in tones:    
        (lowFTIdx, highFIdx)  = tones[number]
        lowFTable, highFTable = F0[lowFTIdx], F1[highFIdx]
        thisError             = np.abs(lowFreq-lowFTable) + np.abs(highFreq-highFTable)
        if thisError < errorMin: 
            identifiedDigit, errorMin = number, thisError
    return identifiedDigit
```
